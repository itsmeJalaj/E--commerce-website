Founded by Sam Walton in 1962, Walmart has grown into one of the largest retailers in the world, serving over 100 million customers worldwide.
            The company is known for its low-price strategy, extensive product selection, and efficient supply chain management. Walmart's business model focuses on cost leadership, ensuring affordability for customers while maintaining profitability. Over the years, it has expanded internationally, adapting to different markets while maintaining its core values of customer satisfaction and accessibility.

```python
## Just by looking at the amount spent by males and females per transaction , we can seee that males spend more than females
## but this is only for the data provided to us , for whole population we need to apply Central Limit Theorem and find the Confidence Intervak
```

```python
male_amt=data[data['Gender']=='M']['Purchase']
male_amt
```

```python
female_amt=data[data['Gender']=='F']['Purchase']
female_amt
```

```python
sns.histplot(data=male_amt ,kde = True )
plt.show()
```

```python
sns.histplot(data=female_amt ,kde = True )
plt.show()
```

```python
## From above two plots we can see that the spending pattern of both the males and females are similiar 
## so we can't say that one gender spends more than the other
## to prove that we apply CENTRAL LIMIT THEOREM
```
